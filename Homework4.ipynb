```python
import random as rnd

def sayiUret(num):
    return rnd.randint(1,num)

try:
  ad=input("Adınız: ")
  print("Hoşgeldin",ad.capitalize(),"hadi başlayalım")

  a=sayiUret(10)
  b=sayiUret(10)
  sor=0
  cevap=0
  while(sor==cevap):
    sor=int(input(str(a)+" * "+str(b)+" = kaçtır "))
    cevap=a*b
    a=cevap
    b=sayiUret(a)
  print("Sanırım bu seni zorladı sevgili ",ad.capitalize())
except ValueError:
  print("Geçersiz karakter girdin")
except TypeError:
  print("Tür yanlış")
except:
  print("Ben sana hiçbir şey demiyorum.")
```
